#### sorting DataFrame or a Series

```python
earthquakes.Magnitude.sort_values(ascending=False)
```

```python
earthquakes.sort_values('Magnitude', ascending=False)  # not affecting underlying df, use inplace=True
```

```python
# sort by multiple columns
earthquakes.sort_values(['MagType', 'Magnitude']) 
```

#### filtering DataFrame by a column value

```python
# short version
earthquakes[earthquakes.Magnitude > 6.5]

# better vesrion
earthquakes.loc[earthquakes.Magnitude > 6.5]
```

```python
# selecting specific columns after filtering

# short version
earthquakes[earthquakes.Magnitude > 6.5].DateTime

# better version
earthquakes.loc[earthquakes.Magnitude > 6.5, 'DateTime']
```

#### filtering by multiple criteria

```python
# & (and) operator
earthquakes[(earthquakes.Magnitude > 6.5) & (earthquakes.MagType == 'Mw')]
```

```python
earthquakes.MagType.unique()
```

```python
# | (or) operator
earthquakes[(earthquakes.MagType == 'Mc') | (earthquakes.MagType == 'Me')]
```

```python
# when filtering df on same series as above, shorter version is:
earthquakes[earthquakes.MagType.isin(['Mc', 'Me'])]
```

#### iterating over a series or a dataFrame

```python
# iterating over a series similar to iterating over a list
for d in earthquakes.Depth.iloc[:5,]:
    print(d)
```

```python
# iterating over dataframe rows:
for index, row in earthquakes.iloc[:5,].iterrows():
    print(index, row.DateTime, row.Magnitude)
```

#### using axis parameter

```python
# axis=0 is rows axis, axis=1 is column axis. 
# e.g. if i want to remove(drop) a column i specify axis=1
# alternatively can specify axis='index'(same as axis=0) or axis='columns' (same as axis=1)
```

```python
# for exampe the default axis parameter for mean() function is axis=0 (i.d. moved down, on row axis)
earthquakes.mean()
```

```python
# explicitly:
earthquakes.mean(axis=0)    
```

```python
# alternative:
earthquakes.mean(axis='index')
```
